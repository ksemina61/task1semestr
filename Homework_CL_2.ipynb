```python
# задание 1
import nltk # импрортирую библиотеку nltk
import re # а также регулярные выражения
from collections import Counter # я импортирую модуль каунтер для того, чтобы считались количество общих слов
from nltk.stem import SnowballStemmer # ну и сноуболстемминг для стемминга
# Для начала нужно почистить текст от ненужных символов и пунктуации, для этого я использую метод clean, еще мне понадобится регулярное выражение
def my_text(text):
  cleaned_text = re.sub(r'[^\w\s]', '', text) #re.sub - ищет шаблон в строке и заменяет его на указанную подстроку. Если шаблон не найден, строка остается неизменной
  cleaned_text = cleaned_text.lower() # привожу все к нижнему регистру, чтобы потом не возникли проблемы
  return cleaned_text
# Теперь мне нужно сделать стемминг, тоже оформлю функцией
def stem_words(words):
  stemmer = SnowballStemmer("english")
  stemmed_words = [stemmer.stem(word) for word in words] # я применяю стеммер ко всем словам в списке words с использованием генератора списка и получаю список основ слов
  return stemmed_words
# Теперь мне нужно классифицировать тексты по темам
def classify_text(filename):
    base_texts = [] # пустой список base_texts, в который будут загружены содержимое текстовых файлов
    with open('Politics.txt', 'r', encoding='utf-8') as f:
        base_texts.append(f.read())
    with open('Justice.txt', 'r', encoding='utf-8') as f:
        base_texts.append(f.read())
    with open('linguistics.txt', 'r', encoding='utf-8') as f:
        base_texts.append(f.read())
    with open('Health_1.txt', 'r', encoding='utf-8') as f:
        base_texts.append(f.read())
    with open('Health_2.txt', 'r', encoding='utf-8') as f:
        base_texts.append(f.read())
# Теперь нужно обработать текст и прочитать его
    with open(filename, 'r', encoding='utf-8') as f:
        new_text = f.read()
    my_new_text = my_text(new_text)
# Токенизирую текст на слова
    base_tokenized_texts = [text.split() for text in base_texts]
    new_tokenized_text = my_new_text.split()
    # Стемминг
    base_tokenized_stemmed_texts = [stem_words(words) for words in base_tokenized_texts] # c помощью уже созданной функцией stem_ words происходит стемминг для каждого списка слов в base_tokenized_texts
    new_tokenized_stemmed_text = stem_words(new_tokenized_text)
     # Считаю общие слова в исходном тексте и в новом
     # Тут я создала  список common_word_counts, где для каждого стеммированного текста из base_tokenized_stemmed_texts считается количество общих слов с новым стеммированным текстом. Тут я использовала множества для подсчета уникальных общих слов методом intersection (пересечение) и затем посчитала их количество с помощью len
    common_word_counts = [
        len(set(base_text).intersection(set(new_tokenized_stemmed_text)))
        for base_text in base_tokenized_stemmed_texts
    ]
# Дальше мне нужно определить тему текста, основываясь на общие слова
    max_common_words_index = common_word_counts.index(max(common_word_counts))
    themes = ['Sleep health of young adults', 'Sleep health in public health agendas', 'Judiciarisation of the mentally ill', 'Associations between the dark tetrad']
    theme = themes[max_common_words_index]

    return theme
```

```python
new_text_filename = 'new_text.txt'
classified_theme = classify_text(new_text_filename)
print("The theme of the text is:", classified_theme)
```

```python
# я прочитала, что можно сделать морфоразбор с помощью pymorphy2, посмотрела, как им пользоваться тут + посмотрела в 4 семинаре https://github.com/pymorphy2/pymorphy2/blob/master/docs/user/guide.rst
!pip install pymorphy2 # задание №2
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
with open ('Anna_Karenina.txt', 'r', encoding = 'utf-8') as file:
  text = file.read()
prepositions = ['в', 'на','под', 'за', 'о', 'перед', 'с', 'перед', 'над','подле', 'между', 'у', 'среди', 'через', 'меж', 'из-за', 'до', 'для', 'без', 'из-под', 'от']
words = text.split() # делю текст на слова
preposition_occurence = {prep: {} for prep in prepositions} # создаю словарь для того, чтобы программа хранила частоты падежей после предлога
for i in range(len(words) -1):
  if words[i].lower() in prepositions: # для удобства привожу слова к нижнему регистру, является ли слово предлогом
    parsed_word = morph.parse(words[i + 1])[0] # с помощтю метода parse слова разбиваются на морфемы
    if parsed_word.tag.case is not None: # проверка на определние падежа у слова
          current_cases = preposition_occurence[words[i].lower()] # создаю переменную, в которой будет содержаться инфо о частотности падежей, доступ будет происходить через words[i].lower()
          if parsed_word.tag.case not in current_cases: # проверка на определение падежа у слова, строка другими словами -  есть ли падеж parsed_word.tag.case в словаре current_cases
            current_cases[parsed_word.tag.case] = 1  # если падеж не был найден в current_cases, то частота = 1
          else:
            current_cases[parsed_word.tag.case] += 1  # но если мы уже видели этот падеж после данного предлога, и теперь нужно увеличить его частоту на 1
# Получаю результат:
for prepositions, cases in preposition_occurence.items():
    print(f"Предлог '{prepositions}':")
    for case, count in cases.items():
        print(f"  - Падеж '{case}': {count} раз")
```
